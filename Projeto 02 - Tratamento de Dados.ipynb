## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
# Unindo os dataframes utilizando o concat, uma vez que os dados possuem as mesmas colunas
pd.concat([investimento1, investimento2], ignore_index=True)
```

```python
# Exibindo número de linhas
print(f"Após juntar os DataFrames temos {pd.concat([investimento1, investimento2], ignore_index=True).shape[0]} linhas")
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
# Criando um novo dataframe sem as linhas duplicadas

investi_novo = pd.concat([investimento1, investimento2], ignore_index=True).drop_duplicates()
```

```python
display(investi_novo)
```

## 4 - Junte o dataframe de investimento com o de cadastro
